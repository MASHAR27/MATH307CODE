```python
# b
import imageio.v2 as imageio
A = imageio.imread('myimg.jpg')
A.shape




```

```python
# c
import numpy as np
import matplotlib.pyplot as plt
if len(A.shape) > 2:
    A = np.mean(A, axis=2).astype(np.uint8)


U, S, Vt = np.linalg.svd(A, full_matrices=False)


k = 1
S_k = np.diag(S[:k])
U_k = U[:, :k]
Vt_k = Vt[:k, :]


A_compressed = U_k.dot(S_k).dot(Vt_k)


plt.imshow(A_compressed, cmap='gray')
plt.axis('off')
plt.title('Compressed Grayscale Image (Rank-1 Approximation)')
plt.show()
```

```python
# d

num_singular_values_to_retain = [1, 5, 10, 20, 50, 100]

for k in num_singular_values_to_retain:

    S_k = np.diag(S[:k])
    U_k = U[:, :k]
    Vt_k = Vt[:k, :]


    A_compressed = U_k.dot(S_k).dot(Vt_k)


    plt.figure()
    plt.imshow(A_compressed, cmap='gray')
    plt.axis('off')
    plt.title(f'Rank-{k} Approximation')
    plt.show()


    original_size = A.size
    compressed_size = U_k.size + S_k.size + Vt_k.size
    compression_ratio = original_size / compressed_size

    print(f"Number of singular values retained: {k}")
    print(f"Compression ratio: {compression_ratio:.2f}")
    print("===========================")
```

```python
# e

k = 6


S_k = np.diag(S[:k])
U_k = U[:, :k]
Vt_k = Vt[:k, :]


A_compressed = U_k.dot(S_k).dot(Vt_k)


plt.figure(figsize=(10, 5))


plt.subplot(1, 2, 1)
plt.imshow(A, cmap='gray')
plt.axis('off')
plt.title('Original Image')


plt.subplot(1, 2, 2)
plt.imshow(A_compressed, cmap='gray')
plt.axis('off')
plt.title(f'Reconstructed Image (k={k})')

plt.tight_layout()
plt.show()
```
